```r
# load data
smokers = read.table('https://www.itl.nist.gov/div898/education/anova/smokers.dat', 
                     header=TRUE, skip=24, col.names = c('Y', 'X1', 'X2'))

smokers.df = data.frame(smokers = smokers$Y, income = smokers$X1, age = smokers$X2)
write.csv(smokers.df, file = '../data/smokers.csv', row.names = FALSE)
smokers.df = read.csv('../data/smokers.csv')
smokers.df
```

```r
model = aov(smokers ~ income * age, data = smokers.df)
model
```

```r
# ANOVA table

summary(model)
```

```r
# Check model assumption for normality
# TODO: what do residuals mean is this how we check the model assumptions for normality?

layout(matrix(c(1, 2, 3, 4), 2, 2)) # optional layout 
plot(model)
```
